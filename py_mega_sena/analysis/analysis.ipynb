## Import Dependencies

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading data

```python
base_file = os.path.join(os.getcwd(), '..', 'data', 'files')
```

```python
df = pd.read_excel(os.path.join(base_file, 'mega_sena_clean_data.xlsx'))
```

```python
df.shape
```

```python
df.head()
```

```python
df['Data'] = pd.to_datetime(df.Data, format="%d/%m/%Y")
df.head()
```

```python
df.tail()
```

```python
# Plotando um gráfico de barras com a frequência de cada número sorteado
plt.figure(figsize=(24, 6))
plt.hist(df['bola 1'], bins=60, range=(1, 60), color='blue', alpha=0.5, label="bola 1")
plt.hist(df['bola 2'], bins=60, range=(1, 60), color='red', alpha=0.5, label="bola 2")
plt.hist(df['bola 3'], bins=60, range=(1, 60), color='green', alpha=0.5, label="bola 3")
plt.hist(df['bola 4'], bins=60, range=(1, 60), color='yellow', alpha=0.5, label="bola 4")
plt.hist(df['bola 5'], bins=60, range=(1, 60), color='purple', alpha=0.5, label="bola 5")
plt.hist(df['bola 6'], bins=60, range=(1, 60), color='orange', alpha=0.5, label="bola 6")
plt.legend()
plt.xlabel('Número')
plt.ylabel('Frequência')
plt.title('Frequência de Números Sorteados na Mega Sena')

plt.show()
```

```python
ball_1 = df['bola 1'].value_counts()
ball_1  = ball_1.reset_index()
ball_1 = ball_1.rename(columns={'bola 1': 'frequencia', 'index': 'dezena'})
ball_1.head(20)
```

```python
ball_2 = df['bola 2'].value_counts()
ball_2  = ball_2.reset_index()
ball_2 = ball_2.rename(columns={'bola 2': 'frequencia', 'index': 'dezena'})
ball_2.head(20)
```

```python
ball_3 = df['bola 3'].value_counts()
ball_3 = ball_3.reset_index()
ball_3 = ball_3.rename(columns={'bola 3': 'frequencia', 'index': 'dezena'})
ball_3.head(20)
```

```python
ball_4 = df['bola 4'].value_counts()
ball_4 = ball_4.reset_index()
ball_4 = ball_4.rename(columns={'bola 4': 'frequencia', 'index': 'dezena'})
ball_4.head(20)
```

```python
ball_5 = df['bola 5'].value_counts()
ball_5 = ball_5.reset_index()
ball_5 = ball_5.rename(columns={'bola 5': 'frequencia', 'index': 'dezena'})
ball_5.head(20)
```

```python
ball_6 = df['bola 6'].value_counts()
ball_6 = ball_6.reset_index()
ball_6 = ball_6.rename(columns={'bola 6': 'frequencia', 'index': 'dezena'})
ball_6.shape
```

```python
balls = pd.concat([ball_1, ball_2, ball_3, ball_4, ball_5, ball_6], axis=0)
balls = balls.reset_index()
balls = balls.drop(columns=['index'])
# balls.sort_values(by=['frequencia'], ascending=False)
balls
```

```python
all_balls = balls['dezena'].value_counts()
all_balls = all_balls.reset_index()
all_balls.head(6)
```

```python
plt.figure(figsize=(24, 6))
plt.hist(balls['frequencia'], bins=60, range=(1, 60), color='purple', alpha=0.6, label="Frequência", rwidth=0.8)
plt.legend()
plt.xlabel('Números')
plt.ylabel('Frequência')
plt.title('Frequência de Números Sorteados na Mega Sena')

plt.show()
```

```python
df2 = df.drop(columns=['Concurso', 'Data'])
```
