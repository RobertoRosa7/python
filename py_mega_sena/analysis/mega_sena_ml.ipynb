```python
import os
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import random
```

```python
base_file = os.path.join(os.getcwd(), '..', 'data', 'files')
```

```python
df = pd.read_excel(os.path.join(base_file, 'mega_sena_clean_data.xlsx'))
# df = pd.read_csv(os.path.join(base_file, 'fake-generate-mega-sena.csv'))
```

```python
df.head()
```

```python
# Criando um array de features
x = df.drop(['Concurso', 'Data'], axis=1).values
x
```

```python
# Criando um array de labels
y = df['bola 1'].values
y
```

```python
# Dividindo os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
X_test.shape
```

```python
# Treinando o modelo
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
# Fazendo previsões
y_pred = model.predict([[31, 38, 39, 41, 43, 49]])
```

```python
# Avaliando a precisão do modelo
accuracy = np.mean(y_pred == y_test)
print('Accuracy: {:.5f}%'.format(accuracy))
```

```python
results = []
for _ in range(0, 10000):
    jogo = random.sample(range(1, 61), 6)
    jogo.sort()
    y_pred = model.predict([jogo])
    accuracy = np.mean(y_pred == y_test)
    print('Accuracy: {:.5f}% {}'.format(accuracy, ' '.join([str(n) for n in jogo])))
    results.append({'rate': accuracy, 'jogo':' '.join([str(n) for n in jogo]) })
```

```python
df = pd.DataFrame(data=results, columns=['rate', 'jogo'])
df = df.sort_values(by=['rate'], ascending=False)
df.head(10)
```
