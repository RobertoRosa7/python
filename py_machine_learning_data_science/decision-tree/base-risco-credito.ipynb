# Árvore de Decisão - Risco de Crédito
classificação: (baixo e alto)

```python
import os
import pickle
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
with open(os.path.join('..', 'file-pkl', 'base-risco-credito.pkl'), 'rb') as f:
    x_base_risco_credito, y_base_risco_credito = pickle.load(f)
    f.close()
```

```python
x_base_risco_credito, y_base_risco_credito
```

```python
base_risco_tree = DecisionTreeClassifier(criterion='entropy', random_state=0)
```

```python
base_risco_tree.fit(x_base_risco_credito, y_base_risco_credito)
```

```python
base_risco_tree.feature_importances_
```

```python
names = ['história', 'dívida', 'garantias', 'renda']
figure, ax = plt.subplots(nrows=1, ncols=1, figsize=(20,6))
a = tree.plot_tree(base_risco_tree, feature_names=names,class_names=base_risco_tree.classes_, filled=True)
```

```python
predict = base_risco_tree.predict([[0,0,1,2], [2,0,0,0]])
predict
```
