```python
import os
import pickle
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
```

```python
with open(os.path.join('..', 'file-pkl', 'base-creditcard.pkl'), 'rb') as f:
    x_credit_training, x_credit_test, y_credit_training, y_credit_test = pickle.load(f)
    f.close()
```

```python
x_credit_training.shape, y_credit_training.shape
```

```python
y_credit_test.shape, x_credit_test.shape
```

```python
base_credit_tree = DecisionTreeClassifier(criterion='entropy', random_state=0)
```

```python
base_credit_tree.fit(x_credit_training, y_credit_training)
```

```python
predict = base_credit_tree.predict(x_credit_test)
predict
```

```python
y_credit_test
```

```python
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(base_credit_tree)
cm.fit(x_credit_training, y_credit_training)
cm.score(x_credit_test, y_credit_test)
```

```python
# recall - algoritmo consegue identificar corretamento
# precision - quando identifica um registro tem precisão
# f1-score - média harmônica entre precision e recall
print(classification_report(y_credit_test, predict))
```

```python
names = ['income', 'age', 'loan']
figure, ax = plt.subplots(nrows=1, ncols=1, figsize=(20,10))
a = tree.plot_tree(base_credit_tree, feature_names=names,class_names=['0', '1'], filled=True)
# figure.savefig('base-credit-tree.png')
```
