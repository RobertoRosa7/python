```python
frames = [nomes_f, nomes_m]
```

```python
nomes = pd.concat(frames)['nome'].to_frame()
nomes.head()
```

```python
# Gerando ID aleatório
import numpy as np

np.random.seed(123)
total_alunos = len(nomes)
total_alunos
```

```python
nomes['id_aluno'] = np.random.permutation(total_alunos) + 1
nomes.sample(5)
```

```python
dominios = ['@dominio.com.br', '@contato.com.br']
nomes['dominio'] = np.random.choice(dominios, total_alunos)
nomes.head()
```

```python
nomes['email'] = nomes.nome.str.cat(nomes.dominio).str.lower()
nomes.head()
```

```python
# Criando cursos
uri ='https://tabela-cursos.herokuapp.com/index.html'
cursos = pd.read_html(uri)
cursos
```

```python
cursos = cursos.rename(columns={'Nome do curso': 'nome_do_curso'})
cursos
```

```python
cursos['id_cursos'] = cursos.index + 1
cursos
```

```python
cursos = cursos.set_index('id_cursos')
cursos
```

```python
# Matriculando os alunos no cursos

nomes['matriculas'] = np.ceil(np.random.exponential(size=total_alunos) * 1.5).astype(int)
nomes
```

```python
todas_matriculas = []
x = np.random.rand(20)
prob = x / sum(x)
```

```python
for index, row in nomes.iterrows():
  id_aluno = row.id_aluno
  matriculas = row.matriculas
  for i in range(matriculas):
    mat = [id_aluno, np.random.choice(cursos.index, p=prob)]
    todas_matriculas.append(mat)

matriculas = pd.DataFrame(todas_matriculas, columns=["id_aluno", "id_curso"])
matriculas
```

```python
matriculas_por_curso = matriculas.groupby(by=["id_curso"]).count().join(cursos['nome_do_curso']).rename(columns={'id_aluno':'quantidade_alunos'})
matriculas_por_curso.head()
```
